```python
X_input = []
x_label = []
y_output = []

for i in range(77):
    x, sr_x = soundfile.read('/Users/cooky/Desktop/data/'+str(i).zfill(2)+'.wav')
    x = np.transpose(x)[0]
    chg = glob.glob('/Users/cooky/Desktop/data/'+str(i).zfill(2)+'-*.wav')[0]
    y, sr_y = soundfile.read(chg)
    y = np.transpose(y)[0]
    if (len(y) != len(x)) or (sr_x != sr_y):
        print('Err ' + str(i))
    else:
        print(i)
        print(x.max())
        print(y.max())
        X_input.extend(x)
        y_output.extend(y)
        for _ in range(len(x)):
            x_label.append(int(chg[-6:-4]))
```

```python
import struct

filename = 'X_input.bin'
with open(filename, 'wb') as file:
    for value in X_input:
        packed_value = struct.pack('f', value)  # float 값을 이진 형식으로 패킹
        file.write(packed_value)
```

```python
filename = 'x_label.bin'
with open(filename, 'wb') as file:
    for value in x_label:
        packed_value = struct.pack('i', value)  # float 값을 이진 형식으로 패킹
        file.write(packed_value)
```

```python
filename = 'y_output.bin'
with open(filename, 'wb') as file:
    for value in y_output:
        packed_value = struct.pack('f', value)  # float 값을 이진 형식으로 패킹
        file.write(packed_value)
```
